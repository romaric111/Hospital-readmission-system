```python
df['change'] = df['change'].replace({'yes': 1, 'no': 0})
df['diabetes_med'] = df['diabetes_med'].replace({'yes': 1, 'no': 0})
df['readmitted'] = df['readmitted'].replace({'yes': 1, 'no': 0})
df
```

```python
num_features = ['time_in_hospital', 'n_lab_procedures', 'n_procedures', 'n_medications', \
                'n_outpatient', 'n_inpatient', 'n_emergency']
num_target = num_features + ['change', 'diabetes_med', 'readmitted']
sns.heatmap(df[num_target].corr(), annot=True, linewidth=.5, fmt=".1f")
```

```python
from sklearn.preprocessing import LabelEncoder

le = dict()
data = pd.DataFrame()

for feat in ['age', 'medical_specialty', 'diag_1', 'diag_2', 'diag_3', 'glucose_test','A1Ctest']:
    le[feat] = LabelEncoder()
    data[feat] = le[feat].fit_transform(df[feat])
    
data
```

```python
for feat in num_target: 
    data[feat] = df[feat]

data   
```

```python
features = data.columns[:-1]
X = data[features]
y = data['readmitted']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier


from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
clf = make_pipeline(StandardScaler(), SVC(kernel = 'rbf', gamma='auto'))
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Logistic Regression
clf = make_pipeline(StandardScaler(), LogisticRegression(random_state=0))
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Random Forest
clf = RandomForestClassifier(max_depth=3, random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

importances = clf.feature_importances_
forest_importances = pd.Series(importances, index=features).sort_values(ascending=False)
# forest_importances = fores

fig, ax = plt.subplots()
forest_importances.plot.bar(ax=ax)
ax.set_title("Feature importances")
# ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

```python
#Gradient Boosting
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, \
                                 max_depth=3, random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# CrÃ©er une liste de modÃ¨les
models = {
    "Random Forest": RandomForestClassifier(max_depth=3, random_state=0),
    "Logistic Regression": LogisticRegression(random_state=0),
    "SVM": SVC(kernel='rbf', gamma='auto'),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=3, random_state=0),
}

# Effectuer la validation croisÃ©e pour chaque modÃ¨le
for name, model in models.items():
    cv_scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    print(f"ModÃ¨le: {name}")
    print(f"Scores de validation croisÃ©e: {cv_scores}")
    print(f"Score moyen: {cv_scores.mean():.2f}")
    print(f"Ã‰cart-type des scores: {cv_scores.std():.2f}")
    print("-" * 30)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import joblib

# Charger les donnÃ©es
data = pd.read_csv('hospital_readmissions.csv')

# Inspecter les valeurs uniques de 'readmitted'
print("Valeurs uniques de 'readmitted' :", data['readmitted'].unique())

# Encoder 'readmitted' si nÃ©cessaire
if data['readmitted'].dtype == 'object':
    data['readmitted'] = data['readmitted'].apply(lambda x: 1 if x == 'Yes' else 0)

# Encodage des colonnes non numÃ©riques restantes
data_encoded = pd.get_dummies(data.drop(columns=['readmitted']), drop_first=True)

# Ajouter la cible 'readmitted' Ã  part
X = data_encoded
y = data['readmitted']

# SÃ©parer les donnÃ©es en ensembles d'entraÃ®nement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# CrÃ©er et entraÃ®ner le modÃ¨le
clf = RandomForestClassifier(max_depth=3, random_state=0)
clf.fit(X_train, y_train)

# Sauvegarder le modÃ¨le correctement
joblib.dump(clf, 'hospital_readmissions_model.pkl')
print("ModÃ¨le rÃ©entraÃ®nÃ© et sauvegardÃ© avec succÃ¨s !")
```
